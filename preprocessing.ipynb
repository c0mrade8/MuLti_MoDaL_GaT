```python
save_path = "/content/drive/MyDrive/RAVDESS/Landmarks_Processed"
os.makedirs(save_path, exist_ok=True)
ldmrk_path="/content/drive/MyDrive/RAVDESS/FacialTracking_Actors_01-24"
for ldmrk_file in os.listdir(ldmrk_path):
  ldmrk_file_path=os.path.join(ldmrk_path,ldmrk_file)
  ldmrk_df=pd.read_csv(ldmrk_file_path)
  #points_df=[]
  x_coords=pd.concat([ldmrk_df.iloc[:,11:67], ldmrk_df.iloc[:,297:365]], axis=1).reset_index(drop=True)
  y_coords=pd.concat([ldmrk_df.iloc[:,67:123], ldmrk_df.iloc[:,365:433]], axis=1).reset_index(drop=True)
  x_np=x_coords.to_numpy()
  y_np=y_coords.to_numpy()
  frames_points_np=np.stack([x_np, y_np], axis=2)
  framess_points=frames_points_np.tolist()
  out_file=os.path.join(save_path, ldmrk_file.replace(".csv", ".pkl"))
  with open(out_file, "wb") as f:
    pickle.dump(framess_points, f)
```
